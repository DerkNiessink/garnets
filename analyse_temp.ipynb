```python
import re
import matplotlib.pyplot as plt
import scienceplots
plt.style.use('science')
```

```python
result_dir = 'results/output_2024-06-07_17-00-59'
```

```python
rows = []
number_pattern = re.compile(r'-?\d+\.\d+E[+-]\d+')
with open(f'{result_dir}/STATIS', 'r') as f:
    for line in f:
        numbers = number_pattern.findall(line)
        rows.append(numbers)
        
print(line.strip())
print(rows)
```

```python
temperature = []
i=3
iteration=[]
while i <len(rows):
    temperature.append(float(rows[i][1]))
    iteration.append(float(rows[i-1][0]))
    i+=12
print(temperature)
print(iteration)
```

```python
plt.figure(figsize=(7, 5))
plt.plot(iteration,temperature)
plt.xlabel('$t (ps)$', fontsize=15)
plt.ylabel('$T (K)$', fontsize=14)
plt.show()
```
