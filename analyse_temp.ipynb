```python
import matplotlib.pyplot as plt
import scienceplots
plt.style.use('science')
import numpy as np

from dlpoly.statis import Statis
```

```python
result_dir = 'data/sim_densvar_4'
statis = Statis(f"{result_dir}/run_1/STATIS")
statis.gen_labels()
print(statis.labels)
```

```python
for i in range(1, 11):

    try:
        statis  = Statis(f"{result_dir}/run_{i}/STATIS")

        time = statis.data[:, 1]
        temp = statis.data[:, 4]
        press = statis.data[:, 29]

        fig, axs = plt.subplots(2, 1, figsize=(6, 5), sharex=True)
        plt.suptitle(f"dt = {round(0.0005 / (1.2**(i-1)), 6)} ps")


        axs[0].plot(time, temp)
        axs[0].set_ylabel('$T \ (K)$', fontsize=14)
        axs[0].set_ylim(800, 1200)


        axs[1].plot(time, press)
        axs[1].set_xlabel('$t \ (ps)$', fontsize=15)
        axs[1].set_ylabel(r'$P \ (\text{Pa})$', fontsize=14)
        axs[1].set_ylim(-10, 10)

        plt.subplots_adjust(hspace=0.05) 
        plt.show()
    except:
        pass
```
