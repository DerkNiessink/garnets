```python
import re
rows = []
number_pattern = re.compile(r'-?\d+\.\d+E[+-]\d+')
with open('STATIS', 'r') as f:
    for line in f:
        numbers = number_pattern.findall(line)
        rows.append(numbers)
        


print(line.strip())
print(rows)
```

```python
temperature = []
i=3
iteration=[]
while i <len(rows):
    temperature.append(float(rows[i][1]))
    iteration.append(float(rows[i-1][0]))
    i+=12
print(temperature)
print(iteration)
```

```python
import matplotlib.pyplot as plt
plt.plot(iteration,temperature)
plt.xlabel('time simulation')
plt.xlabel('temperature')
plt.show()
```
