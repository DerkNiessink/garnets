```python
import matplotlib.pyplot as plt
import scienceplots
plt.style.use('science')
import numpy as np

from dlpoly.statis import Statis
```

```python
result_dir = 'results/output_2024-06-07_19-00-47'
statis = Statis(f"{result_dir}/STATIS")
statis.gen_labels()
print(statis.labels)
```

```python
time = statis.data[:, 1]
temp = statis.data[:, 4]
press = statis.data[:, 29]
```

```python
fig, axs = plt.subplots(2, 1, figsize=(6, 5), sharex=True)

axs[0].plot(time, temp)
axs[0].set_ylabel('$T \ (K)$', fontsize=14)

axs[1].plot(time, press)
axs[1].set_xlabel('$t \ (ps)$', fontsize=15)
axs[1].set_ylabel(r'$P \ (\text{Pa})$', fontsize=14)
axs[1].set_ylim(-10, 10)

plt.subplots_adjust(hspace=0) 
plt.show()
```
